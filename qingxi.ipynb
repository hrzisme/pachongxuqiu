```python
import pandas as pd
import jieba
import jieba.analyse

total_str = ''
```

```python
detail_list = []
fname_list = ['java后端_1.csv',
              'java后端_2.csv',
              'java后端_3.csv',
              'java后端_4.csv',
              'java后端_5.csv']

for item in fname_list:
    df = pd.read_csv(item)
    df = df['job_req']
    for item in df:
        print(item)
        detail_list.append(item)
```

```python
with open('stop_words.txt','r+',encoding='utf-8') as f:
    stopword = f.read().split("\n")
```

```python
jieba.load_userdict("userdic.txt")
for i in detail_list:
    sentence_seged = jieba.cut(str(i).strip())
    outstr = ''
    for word in sentence_seged:
        if word not in stopword:
            if word != '\t':
                outstr += word
                outstr += " "
    total_str += outstr
```

```python
# 使用jieba自带分析工具进行tfidf分析
total_str = total_str.lower()
rst = jieba.analyse.extract_tags(total_str,topK=100)
print(rst)
```

```python
from PIL import Image
from wordcloud import WordCloud
import numpy as np
```

```python
words_dict = {}
for i in detail_list:
    temp = jieba.cut(str(i).strip())
    for t in temp:
        if t in stopword or t == '\t' or t ==' ':
            continue
        if t in words_dict.keys():
            words_dict[t] += 1
        else:
            words_dict[t] = 1
```

```python
images = Image.open("123.jpg")  # 打开保存的图片
maskImages = np.array(images)  # 并用numpy转换
wc = WordCloud(font_path="msyh.ttc",
               mask=maskImages,
               background_color="white")\
    .generate(total_str)  # 生成词云图
wc.to_file('wordCloudPi4.png')  # 保存到本地图片文件
```
